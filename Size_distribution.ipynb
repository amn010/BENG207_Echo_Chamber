```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

# ── Settings — edit these ────────────────────────────────────────────────────

FILE   = "BariumChloride-MMT all stacks.xlsx"      # path to your file (.xlsx, .xls, .csv, .tsv)
COLUMN = "Length"       # column name to plot
BIN_SIZE = 10          # width of each bin (in data units)
x_min = 0
X_MAX = 180
TITLE  = "2.5% BaCl\u2082/2.5% CaCl\u2082  With MMT Particle Size Distribution -- Batch 1"   # plot title
COLOR  = "green"      # bar colour
OUTPUT = "BaCl2-MMT_size_batch1.png"          # set to e.g. "plot.png" to save, or None to display

# ── Load data ────────────────────────────────────────────────────────────────

ext = FILE.rsplit(".", 1)[-1].lower()
if ext in ("csv",):
    df = pd.read_csv(FILE)
elif ext == "tsv":
    df = pd.read_csv(FILE, sep="\t")
else:
    df = pd.read_excel(FILE)  # handles .xlsx, .xls, .ods automatically

data = pd.to_numeric(df[COLUMN], errors="coerce").dropna().to_numpy()

# ── Plot ─────────────────────────────────────────────────────────────────────

weights = np.ones(len(data)) / len(data) * 100   # convert counts → percentages

fig, ax = plt.subplots(figsize=(9, 5))

x_max = X_MAX if X_MAX is not None else data.max()
bin_edges = np.arange(0, x_max + BIN_SIZE, BIN_SIZE)
_, bin_edges, _ = ax.hist(data, bins=bin_edges, weights=weights,
                           color=COLOR, edgecolor="white", linewidth=0.6)

ax.set_xlabel("Particle Diameter (\u03bcm)", fontsize=12)
ax.set_ylabel("Frequency (%)", fontsize=12)
ax.set_title(TITLE, fontsize=14, fontweight="bold", pad=12)

ax.set_xticks(np.arange(x_min, x_max + 10, 10))                  # start at 0, step by 10
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{int(x):,}"))
ax.yaxis.set_major_formatter(mticker.FormatStrFormatter("%.1f%%"))
plt.xticks(rotation=0, fontsize= 12)

ax.patches[0].set_label(f"Bin size: {BIN_SIZE}")
ax.legend(fontsize=10)

ax.yaxis.grid(True, linestyle="--", alpha=0.5, zorder=0)
ax.set_axisbelow(True)
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()

if OUTPUT:
    fig.savefig(OUTPUT, dpi= 300)
    print(f"Saved to {OUTPUT}")
else:
    plt.show()
```
